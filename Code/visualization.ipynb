# Visualisation

### Preperation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/kaggle/input/rating2/Textrating.xlsx")
df.head()
```

```python
# Check for missing values
df.isna().sum()
```

### Generate figure 1

```python
# Mean values of data quality dimensions by AI
figure_1 = df[["Model","Relevance","Accuracy","Consistency","Completeness"]].groupby("Model").mean()
figure_1
```

```python
# Create plot

figure_1.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Model', fontsize = 10)
plt.ylabel('Rating', fontsize = 12)
#plt.title('Data Quality Dimensions by Model', fontsize=14)
plt.legend(fontsize = 15)
plt.xticks(fontsize = 12, rotation=0)
plt.ylim((1,5))
plt.yticks(fontsize = 12)
plt.show()
```
